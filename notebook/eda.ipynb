# Dataset Understanding 📊 

This dataset represents user clickstream behavior in an e-commerce clothing store (2008).

Each row corresponds to a user interaction within a session, including:

- Product viewed
- Page location
- Navigation order
- Product attributes (price, color, category)

**The dataset enables analysis of:**

- User navigation patterns
- Product preferences
- Impact of UI placement on behavior
- Differences across countries

### Insight: Dataset Quality

The dataset contains 165,474 user interactions across 24,026 sessions.

There are no missing values, indicating a clean and reliable dataset for analysis

Most variables are numerical, with only one categorical feature (clothing model)

### Insight: User Engagment

Most user sessions are relatively short, with a median of 4 interactions per session.

However, there are extreme cases with up to 195 clicks, indicating the presence of highly engaged users.

This suggests a skewed distribution where a small group of users contributes significantly to overall interaction


### Insight: Product Category Preference

Trousers (category 1) receive the highest number of interactions, making them the most popular product category

Sales item (category 4) also show strong engagement, suggesting that discounts significantly influence user behavior

Other categories (skirts and blouses) have similar interaction levels, indicating balanced interest

### Insight: Impact of Product Placement

Products displayed in the top-left and top-middle sections receive the highest number of interactions

In contrast, items located in the bottom-right area receive the least attention

This indicates that visibility and position on the page strongly influece user engagement, highlighting the importance of UI design in e-commerce plarforms
